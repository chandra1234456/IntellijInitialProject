# Coroutines

 launch is an extension function on CoroutineScope. In a normal function like main(), there is no scope available. We must use runBlocking or provide a CoroutineScope to start a coroutine.

launch {
    delay(1000)
    println("Hello")
}

  Why does "Done" print before "Hello"?
    Because launch is asynchronous. The main thread continues execution immediately without waiting.

```kotlin
viewModelScope.launch {
    async {
        delay(1000)
        println("API Call")
    }
}
```

Is this correct usage of async?
What is the problem here?
Will "API Call" print?
What is the correct way to write it?

async is used for returning a value (Deferred<T>)
Here, you're NOT calling .await()
So using async is unnecessary
It behaves like launch but creates a Deferred that you ignore

```kotlin
viewModelScope.launch {
    delay(1000)
    println("API Call")
}
viewModelScope.launch {
    launch {
        delay(1000)
        println("API Call")
    }
}
```

What happens if we don’t call await()?

The coroutine still executes, but the Deferred result is ignored. This may lead to unhandled exceptions or wasted memory

Thread.sleep() blocks the thread

delay() suspends the coroutine without blocking

viewModelScope.launch(Dispatchers.Main) {
    val result = withContext(Dispatchers.Main) {
        Thread.sleep(3000)
        "Done"
    }
    println(result)
}

Thread.sleep blocks the main thread when used on Dispatchers.Main, causing UI freeze and possible ANR. Heavy or blocking work must be moved to Dispatchers.IO or Default using withContext. Also, delay() should be preferred over Thread.sleep() because it suspends without blocking.

Does delay block the thread?
No. delay suspends the coroutine without blocking the underlying thread.

What is difference between supervisorScope and SupervisorJob?
supervisorScope → scoped supervision (local)

SupervisorJob → applied at CoroutineScope level (global behavior)
