### 1. What is Git?
Git is a **distributed version control system** used to track changes in source code and collaborate with others.

---

### 2. What is Version Control?
Version control helps manage changes to files over time, allowing you to:
- Track history
- Revert changes
- Collaborate safely

---

### 3. Difference between Git and GitHub?
| Git | GitHub |
|---|---|
| Version control tool | Hosting platform |
| Works locally | Works online |
| Tracks code changes | Stores & shares Git repositories |

---

### 4. What is a repository?
A repository (repo) is a place where project files and their version history are stored.

---
