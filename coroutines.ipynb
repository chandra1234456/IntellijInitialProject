ğŸ¯ 2. Coroutine Interview Q&A (With Answers)
ğŸ”¹ Basics
Q1: What is a Coroutine?

Answer:
A coroutine is a lightweight thread used for asynchronous programming. It allows writing async code sequentially without blocking threads.

Q2: Difference between launch and async?

Answer:

launch	async
Returns Job	Returns Deferred<T>
Used for fire-and-forget	Used when result is needed
Exception crashes parent	Exception returned on await()
Q3: What is a suspend function?

Answer:
A function that can pause and resume execution without blocking the thread.

ğŸ”¹ Dispatchers & Scopes
Q4: What are Dispatchers?

Answer:
They decide which thread the coroutine runs on.

Main â†’ UI

IO â†’ Network, DB

Default â†’ CPU work

Q5: Why avoid GlobalScope?

Answer:
It is not lifecycle-aware and causes memory leaks.

ğŸ”¹ Structured Concurrency
Q6: What is structured concurrency?

Answer:
Child coroutines are bound to the parent scope and automatically canceled when the parent is canceled.

Q7: Difference between coroutineScope and supervisorScope?

Answer:

coroutineScope: Failure cancels all children

supervisorScope: Failure does NOT cancel siblings

ğŸ”¹ Exception Handling
Q8: Why try/catch doesnâ€™t work with launch?

Answer:
Because launch doesnâ€™t return a result; exceptions are thrown immediately.

Q9: How to handle exceptions in async?

Answer:
Wrap await() inside try/catch.

ğŸ”¹ Android Lifecycle
Q10: What happens to viewModelScope when ViewModel is cleared?

Answer:
All coroutines are automatically canceled.

ğŸ”¹ Flow
Q11: Difference between Flow and LiveData?

Answer:

Flow â†’ Cold, supports operators

LiveData â†’ Lifecycle-aware but limited

Q12: StateFlow vs SharedFlow?

Answer:
